### Stop Words Removal

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
sentence="this is a sample sentence, showing of the stop words filterization"
```

```python
words=word_tokenize(sentence)
```

words

```python
tokens=word_tokenize(sentence.lower())
```

```python
new_senetence=[word for word in tokens if word in stopwords.words('english')]
```

```python
new_senetence
```

```python
new_senetence=[word for word in tokens if word not in stopwords.words('english')]
```

```python
new_senetence
```
